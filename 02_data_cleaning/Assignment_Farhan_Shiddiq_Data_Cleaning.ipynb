1. Lakukan handling outlier pilih 1 kolom saja yang di handling dengan IQR setelah mengecek distribusinya
2. Dan cek distribusinya kembali

```python
def check_plot(df, variable):

    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df[variable], bins=30)
    plt.title('Histogram')

    # plot Q-Q
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('Variable quantiles')

    # box plot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')

    plt.show()
```

```python
# memvisualisasikan outlier
check_plot(data_clf, 'MedInc')
```

```python
# Handling Outlier
Q1 = df_california.copy()['MedInc'].quantile(0.25)
Q3 = df_california.copy()['MedInc'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

### untuk exploratory data analysis

* Berdasarkan visualisasi histogram, probability plot, dan boxplot, kolom MedInc menunjukkan distribusi yang tidak normal dan cenderung skewed ke kanan. Mayoritas data berada pada rentang nilai 2 hingga 5, namun terdapat beberapa nilai ekstrem hingga di atas 14.

* Probability plot menunjukkan penyimpangan signifikan dari garis normalitas, terutama pada bagian kanan distribusi. Hal ini mengindikasikan bahwa data tidak mengikuti distribusi normal.

* Boxplot juga memperlihatkan banyak outlier pada bagian atas, yang menunjukkan adanya nilai pendapatan median yang sangat tinggi dibandingkan mayoritas data.

* Karena adanya outlier yang cukup banyak dan distribusi yang tidak normal, dilakukan handling outlier menggunakan metode IQR untuk mengurangi pengaruh nilai ekstrem terhadap analisis statistik selanjutnya.

# C. Missing Value Handling
Missing Value Merupakan nilai yang hilang di suatu kolom, namun tidak hanya sebatas itu. Untuk nilai yang tidak sesuai tapi muncul didalam suatu kolom tertentu juga bisa dianggap sebagai missing value. Misal tipenya kategorikal tapi isinya ada yang berupa angka dan juga sebaliknya.

### Aturan missing value handling :
1. jika tipe datanya itu **numerik** -> handling menggunakan **median** (robust / tahan terhadap outlier)
2. jika tipe datanya itu **kategorik / object / string** -> handling menggunakan **mode / modus**

**Check Type Dataset**

```python
print("California")
print(df_california.dtypes)
print("----------------------------")
print("Telco Customer Churn")
print(df_tcc.dtypes)
print("----------------------------")
print("Titanic")
print(df_titanic.dtypes)
print("----------------------------")
print("Company")
print(df_company.dtypes)
```

Memisahkan Kolom Numerik dan Kategorikal

```python
# California
numeric_cols = df_california.select_dtypes(include=['int64','float64']).columns
categorical_cols = df_california.select_dtypes(include=['object']).columns
# Telco Customer Churn
numeric_cols = df_tcc.select_dtypes(include=['int64','float64']).columns
categorical_cols = df_tcc.select_dtypes(include=['object']).columns
# Titanic
numeric_cols = df_titanic.select_dtypes(include=['int64','float64']).columns
categorical_cols = df_titanic.select_dtypes(include=['object']).columns
# Company
numeric_cols = df_company.select_dtypes(include=['int64','float64']).columns
categorical_cols = df_company.select_dtypes(include=['object']).columns
```

Handling Kolom Numerik

```python
# Numerik
dataset = [df_california, df_tcc, df_titanic, df_company]

for df in dataset:
    numeric_cols = df.select_dtypes(include=['int64','float64']).columns
    for col in numeric_cols:
        df[col] = df[col].fillna(df[col].median())
```

Kategorikal

```python
# Categorical
dataset = [df_california, df_tcc, df_titanic, df_company]

for df in dataset:
  categorical_col = df.select_dtypes(include=['object']).columns
  for col in categorical_col:
    df[col] = df[col].fillna(df[col].mode()[0])
```

1. One Hot Encoding (OHE) : Lakukan pada kolom gender, dan anda eksplor potensi kolom lainnya
